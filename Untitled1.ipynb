```python
import numpy as np 
#create a 1-D array by passing a list into numpy is array() function. 
print(np.array([8,4,6,0,2]))
```

```python
print ('create a 2-D array by passing a list of lists into array().')
A= np.array ([ [1,2,3],[4,5,6] ])
print(A)

print('Access elements of the array with brackets .')
print(A[0,1], A[1,2])

print('The elements of a 2-D array are 1-D arrays.')
print(A[0])
```

```python
print('Addition concatentes list together')
print([1,2,3] + [4,5,6])

print('Mutliplication concatenates a list with itself a given number of tiems')
print([1,2,3] * 4)
```

```python
x = np.arange(0, 50, 10)
index = np.array([3, 1 , 4])
print(x[index])
#A boolean array extracts the elements of 'x'at the same places as 'True'.
mask = np.array([True,False,False,True,False])
print(x[mask])
```

```python
#Fancy indexing vith condition
y = np.arange(10,20,2)#Every other integers from 10 to 20.
print(y)
mask =y >15 #Same as np.arrar ([i > 15 for i in y]).
print(mask)
print(y[mask])

#Change the values of 'y'that are larger than 15 to 100.
y[mask] = 100
print(y)
```
